Importing the dependencies (labels and functions)

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import nltk
from nltk.stem import PorterStemmer
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Stopwords do not add much value to the dataset
print(stopwords.words('english'))
```

Data Pre-processing

```python
# Loading the dataset to Pandas dataframe
# Pandas dataframe loads the data in a more structured table format
news_dataset = pd.read_csv('/content/train.csv')
```

```python
news_dataset.shape
```

```python
# Printing the first 5 rows of the dataframe
news_dataset.head()
```

```python
# Counting the number of missing values in the dataset
# The number of missing values represent the number of missing values in each column of the datset
# The values might be missing because it was not available in the first place while preparing the dataset
news_dataset.isnull().sum()
```

```python
# Replacing the null values with empty string
news_dataset = news_dataset.fillna('')
```

```python
# Merging the News Title/Headlines column with Authors column
news_dataset['content'] = news_dataset['author']+' '+ news_dataset['title']
print(news_dataset['content'])
```

```python
# Seperating the data in the content column and label column
# axis parameter helps in removing a column in the dataset
X = news_dataset.drop(columns='label', axis=1)
Y = news_dataset['label']
print(X)
print(Y)
```

Stemming:
It is the process of reducing a word to its root word.

```python
port_stem = PorterStemmer()
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ', content)   # Removes all the other values except alphabets and replaces the other values with an empty string/space
  stemmed_content = stemmed_content.lower() # Converting all the alphabets to lowercase
  stemmed_content = stemmed_content.split() # All the words or letters are split and assigned to lists
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')] # Stemming
  stemmed_content = ' '.join(stemmed_content) # All the stemmed words are joined
  return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
print(news_dataset['content'])
```

```python
# Seperating the stemmed content column and the label column
X = news_dataset['content'].values
Y = news_dataset['label'].values
print(X)
print(Y)
```

```python
# Converting textual data to numerical data
# Tf - Term Frequency, detects the repetitive words and assigns an unique numerical value
# idf - inverse document frequency, detects highly repeated words which is of not much value and reduces its importance value
vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X = vectorizer.transform(X)
print(X)
```

Splitting the dataset into training data and testing data


```python
# stratify - splits the data in equal/similar proportions as in the original dataset
# random_state - determines the manner of the data that is being split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

Training the model : Logistic Regression Model

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Evaluation

```python
# Accuracy score of the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data :', training_data_accuracy)
```

```python
# Accuracy score of the testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the testing data :', test_data_accuracy)
```

Creating a Predictive system

```python
X_new = X_test[0]
prediction = model.predict(X_new)
print(prediction)
if (prediction[0]==0):
  print("The news is real")
else:
  print("The news is fake")
```
